```python
#1.Discuss string slicing and provide examples.


##Ans.String slicing is a fundamental concept in programming that allows you to extract a subset of characters from a string. It's a powerful technique used in various programming languages, including Python, JavaScript, and many others.

Basic Syntax:

The basic syntax for string slicing is:

string[start:stop:step]

- start: The starting index of the slice (inclusive).
- stop: The ending index of the slice (exclusive).
- step: The increment between indices (default is 1).

Examples:

1. Simple Slice:

my_string = "Hello World"
print(my_string[0:5])  # Output: "Hello"

Extracts the first 5 characters from the string.

1. Slice with Step:

my_string = "abcdefg"
print(my_string[0:6:2])  # Output: "ace"

Extracts every other character from the string, starting from the first character.

1. Negative Indices:

my_string = "Hello World"
print(my_string[-5:])  # Output: "World"

Extracts the last 5 characters from the string.

1. Omitting Start/Stop:

my_string = "Hello World"
print(my_string[:5])  # Output: "Hello"
print(my_string[6:])  # Output: "World"

Omitting the start index starts the slice from the beginning, while omitting the stop index goes until the end of the string.

1. Reversing a String:

my_string = "Hello World"
print(my_string[::-1])  # Output: "dlroW olleH"

Extracts the entire string in reverse order.

These examples demonstrate the versatility of string slicing in extracting specific parts of a string. The syntax may vary slightly depending on the programming language, but the concepts remain similar.


```

```python
#2.Explain the key features of lists in Python.

##Ans.In Python, lists are a fundamental data structure that offer a versatile way to store and manipulate collections of items. Here are the key features of lists in Python:

1. Ordered Collection: Lists are ordered, meaning that the order of the elements matters and is preserved.

2. Mutable: Lists are mutable, which means you can modify their contents after creation (e.g., add, remove, or change elements).

3. Indexing: Lists support indexing, allowing you to access specific elements by their position (index) in the list.

4. Slicing: Lists support slicing, enabling you to extract subsets of elements using a start:stop:step syntax.

5. Dynamic Size: Lists can grow or shrink dynamically as elements are added or removed.

6. Heterogeneous: Lists can store elements of different data types, including strings, integers, floats, and other lists.

7. Nested Lists: Lists can contain other lists, enabling the creation of complex data structures like matrices or trees.

8. Methods: Lists have various built-in methods for common operations like:
    - append(): Add an element to the end.
    - insert(): Insert an element at a specific position.
    - remove(): Remove the first occurrence of an element.
    - sort(): Sort the list in-place.
    - reverse(): Reverse the list in-place.

9. List Comprehensions: Python supports list comprehensions, a concise way to create lists from existing iterables or expressions.

10. Iterability: Lists are iterable, allowing you to loop through their elements using for loops or other iteration constructs.

These features make lists an essential tool in Python programming, suitable for a wide range of tasks, from simple data storage to complex data manipulation and processing.
```

```python
#3.Describe how to access, modify and delete elements in a list with examples.

##Ans.Here's a step-by-step guide on how to access, modify, and delete elements in a list with examples:

Accessing Elements:

1. Indexing: Use square brackets [] with the index of the element you want to access.

my_list = [1, 2, 3, 4, 5]
print(my_list[0])  # Output: 1
print(my_list[4])  # Output: 5

1. Negative Indexing: Use negative indices to access elements from the end of the list.

print(my_list[-1])  # Output: 5
print(my_list[-5])  # Output: 1

1. Slicing: Use square brackets [] with a start:stop:step syntax to access a subset of elements.

print(my_list[1:3])  # Output: [2, 3]
print(my_list[1:5:2])  # Output: [2, 4]

Modifying Elements:

1. Assignment: Use indexing to assign a new value to an element.

my_list[0] = 10
print(my_list)  # Output: [10, 2, 3, 4, 5]

1. Insertion: Use the insert() method to add an element at a specific position.

my_list.insert(2, 7)
print(my_list)  # Output: [10, 2, 7, 3, 4, 5]

1. Append: Use the append() method to add an element to the end of the list.

my_list.append(6)
print(my_list)  # Output: [10, 2, 7, 3, 4, 5, 6]

Deleting Elements:

1. Assignment: Use indexing to delete an element by assigning None or an empty value.

my_list[0] = None
print(my_list)  # Output: [None, 2, 7, 3, 4, 5, 6]

1. Remove: Use the remove() method to delete the first occurrence of an element.

my_list.remove(7)
print(my_list)  # Output: [None, 2, 3, 4, 5, 6]

1. Pop: Use the pop() method to delete an element at a specific position (default is the last element).

my_list.pop(0)
print(my_list)  # Output: [2, 3, 4, 5, 6]

1. Del: Use the del statement to delete an element or a slice of elements.

del my_list[0]
print(my_list)  # Output: [3, 4, 5, 6]

Remember to handle index errors and out-of-range indices when accessing or modifying elements in a list.
```

```python
#4.Compare and contrast tuples and lists with examples.

##Ans.Tuples and lists are both data structures in Python used to store collections of items. Here's a comparison of their key differences and similarities:

Similarities:

1. Both can store multiple items of different data types.
2. Both support indexing, slicing, and iterating.
3. Both can be nested (contain other tuples or lists).

Differences:

1. Immutability: Tuples are immutable (cannot be changed after creation), while lists are mutable (can be modified after creation).
2. Syntax: Tuples use parentheses () , while lists use square brackets [].
3. Performance: Tuples are faster and more memory-efficient than lists due to their immutability.
4. Use Cases: Tuples are suitable for storing data that shouldn't be changed, like co-ordinates or dates, while lists are suitable for storing data that needs to be modified, like user input or dynamic data.

Examples:

Tuples:

my_tuple = (1, 2, 3, 4, 5)
print(my_tuple[0])  # Output: 1
# Trying to modify a tuple will raise an error
my_tuple[0] = 10  # TypeError: 'tuple' object does not support item assignment

Lists:

my_list = [1, 2, 3, 4, 5]
print(my_list[0])  # Output: 1
my_list[0] = 10  # Modifying a list is allowed
print(my_list)  # Output: [10, 2, 3, 4, 5]

In summary, tuples are ideal for storing immutable data, while lists are suitable for storing mutable data. Understanding the differences between tuples and lists can help you choose the right data structure for your specific use case.
```

```python
#5.Describe the key features of sets and provide examples of their use.

##Ans.Sets are a fundamental data structure in Python, offering a unique way to store and manipulate collections of unique elementsHere are the key features of sets and examples of their use:

Key Features:

1. Unordered Collection: Sets are unordered, meaning that the order of elements doesn't matter.
2. Unique Elements: Sets only store unique elements, automatically removing duplicates.
3. Mutable: Sets are mutable, allowing you to add or remove elements after creation.
4. Fast Membership Testing: Sets provide fast membership testing using the in operator.
5. Set Operations: Sets support various set operations like union, intersection, difference, and symmetric difference.

Examples:

1. Creating a Set:

my_set = {1, 2, 3, 4, 5}
print(my_set)  # Output: {1, 2, 3, 4, 5}

1. Adding Elements:

my_set.add(6)
print(my_set)  # Output: {1, 2, 3, 4, 5, 6}

1. Removing Elements:

my_set.remove(4)
print(my_set)  # Output: {1, 2, 3, 5, 6}

1. Union and Intersection:

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2))  # Output: {1, 2, 3, 4, 5}
print(set1.intersection(set2))  # Output: {3}

1. Set Comprehensions:

my_set = {x**2 for x in range(5)}
print(my_set)  # Output: {0, 1, 4, 9, 16}

Use Cases:

1. Removing Duplicates: Sets automatically remove duplicates, making them ideal for tasks like data cleaning.
2. Fast Membership Testing: Sets provide fast membership testing, making them suitable for tasks like data validation.
3. Set Operations: Sets support various set operations, making them useful for tasks like data analysis and filtering.
4. Data Analysis: Sets can be used to perform various data analysis tasks like finding unique elements, intersections, and unions.

In summary, sets are a powerful data structure in Python, offering a unique combination of features that make them ideal for tasks involving unique elements, fast membership testing, and set operations.
```

```python
#7.Describe how to add, modify and delete items in a dictionary with examples. 

##Ans.Dictionaries in Python are mutable data structures that store mappings of unique keys to values. Here's how to add, modify, and delete items in a dictionary:

Adding Items:

1. Direct Assignment: Assign a value to a new key using the syntax dict[key] = value.


my_dict = {}
my_dict['name'] = 'John'
print(my_dict)  # Output: {'name': 'John'}


1. update() Method: Update the dictionary with new key-value pairs using the update() method.


my_dict = {'name': 'John'}
my_dict.update({'age': 30, 'city': 'New York'})
print(my_dict)  # Output: {'name': 'John', 'age': 30, 'city': 'New York'}


Modifying Items:

1. Direct Assignment: Modify an existing value by assigning a new value to its key.


my_dict = {'name': 'John', 'age': 30}
my_dict['age'] = 31
print(my_dict)  # Output: {'name': 'John', 'age': 31}


1. update() Method: Update the value of an existing key using the update() method.


my_dict = {'name': 'John', 'age': 30}
my_dict.update({'age': 31})
print(my_dict)  # Output: {'name': 'John', 'age': 31}


Deleting Items:

1. del Statement: Remove a key-value pair using the del statement.


my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
del my_dict['city']
print(my_dict)  # Output: {'name': 'John', 'age': 30}


1. pop() Method: Remove a key-value pair and return the value using the pop() method.


my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
removed_value = my_dict.pop('city')
print(my_dict)  # Output: {'name': 'John', 'age': 30}
print(removed_value)  # Output: 'New York'


1. clear() Method: Remove all key-value pairs using the clear() method.


my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
my_dict.clear()
print(my_dict)  # Output: {}
```

```python
#8.Discuss the importance of dictionary keys being immutable and provide examples. 

##Ans.In Python, dictionary keys must be immutable because:

1. Hashability: Dictionary keys are used to calculate a hash value, which is used to store and retrieve values efficiently. Immutable objects can be hashed, whereas mutable objects cannot.
2. Uniqueness: Immutable keys ensure that each key is unique and cannot be changed, preventing accidental overwriting of values.
3. Efficient Lookups: Immutable keys enable fast lookups, as the hash value remains constant.

Examples of immutable objects that can be used as dictionary keys:

1. Strings: "hello", 'hello'
2. Integers: 1, 2, 3
3. Floats: 3.14, -0.5
4. Tuples: (1, 2), ("a", "b")

Examples of mutable objects that cannot be used as dictionary keys:

1. Lists: [1, 2], ["a", "b"]
2. Dictionaries: {"a": 1}, {"b": 2}
3. Sets: {1, 2}, {"a", "b"}

Trying to use a mutable object as a dictionary key will raise a TypeError:


my_dict = {[1, 2]: "value"}  # TypeError: unhashable type: 'list'


To use a mutable object as a key, convert it to an immutable type, like a tuple:


my_dict = {(1, 2): "value"}  # Valid
```
